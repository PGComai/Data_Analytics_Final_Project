```python
# Importing dependecies 
import pandas as pd
```

```python
# Read the first CSV file into a dataframe
df1 = pd.read_csv("Bx_Books/Cleaned_BXBOOKS.csv", encoding='ISO-8859-1')
```

```python
df1.columns
```

```python
# Read the second CSV file into a dataframe
df2 = pd.read_csv("Bx_Books/Cleaned_BX_Ratings.csv", encoding='ISO-8859-1')
```

```python
df2.columns
```

```python
# Read the second CSV file into a dataframe
df3 = pd.read_csv("Bx_Books/Cleaned_BX-Users.csv", encoding='ISO-8859-1')
```

```python
df3.columns
```

```python
# Merge the two dataframes on a common column (e.g. "ISBN")
merged_df1 = pd.merge(df1, df2, on="ISBN")
```

```python
merged_df1.columns
```

```python
# Merge the two dataframes on a common column (e.g. "User-ID")
BX_Books = pd.merge(merged_df1, df3, on="User-ID")
```

```python
BX_Books
```

```python
# Check if there are any duplicates in the DataFrame
duplicates = BX_Books.duplicated().sum()

# Display the number of duplicates
print("Number of duplicates:", duplicates)
```

```python
# Verify that there are no more missing values
null = BX_Books.isna().sum().sum()
print("Number of missing values:", null)
```

```python
# Get a list of the current column names
columns = list(BX_Books.columns)
columns
```

```python
# Find the index of the column that you want to move
column_index = columns.index("User-ID")
column_index
```

```python
# Remove the column from the list of columns
columns.pop(column_index)
```

```python
# Insert the column at a new location in the list of columns
columns.insert(0, "User-ID")
```

```python
# Reorder the columns in the dataframe
BX_Books = BX_Books.reindex(columns=columns)
```

```python
# Verify DataFrame
BX_Books
```

```python
BX_Books.rename(columns={'User-ID': 'User_ID', 'Location': 'Address', 'Book-Rating': 'Book_Rating', 'Book-Title': 'Book_Title', 'Book-Author': 'Book_Author', 'Year-Of-Publication': 'Year_Of_Publication'}, inplace=True)
```

```python
BX_Books
```

```python
# Saving df as a DA_fianls csv
BX_Books.to_csv("DA_finals.csv", index=False)
```
