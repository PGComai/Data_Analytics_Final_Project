```python
import pandas as pd

df = pd.read_csv("../CSV/BC_Final_clean_again.csv")
df.head()
```

```python
title_isbn_df = df[['ISBN', 'Book_Title']].copy()
title_isbn_df.head()
```

```python
isbns = title_isbn_df.groupby('ISBN')
isbns.count()
```

```python
titles = title_isbn_df.groupby('Book_Title')
titles.count()
```

```python
isbn_lut = isbns.head(1).set_index('ISBN').to_dict(orient='index')

isbn_lut
```

```python
renamed_df = df.copy()

renamed_df['Book_Title'] = renamed_df['ISBN'].apply(lambda x: isbn_lut[x]['Book_Title'])
renamed_df.head()
```

```python
r_title_isbn_df = renamed_df[['ISBN', 'Book_Title']].copy()
r_title_isbn_df.head()
```

```python
r_isbns = r_title_isbn_df.groupby('ISBN')
r_isbns.count()
```

```python
r_titles = r_title_isbn_df.groupby('Book_Title')
r_titles.count()
```

```python
c = df['Book_Title'] == renamed_df['Book_Title']

c.value_counts()
```

```python
# turns out the ISBN and Book_Title columns were perfectly correlated
```
