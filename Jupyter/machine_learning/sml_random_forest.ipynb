```python
# Initial imports.
import pandas as pd
from pathlib import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Load Data Set
file_path = "/Users/nataliebernal/Desktop/git/Data_Analytics_Final_Project/CSV/countries_binary_encoded.csv"
users_df = pd.read_csv(file_path)
users_df
```

```python
users_df = users_df.drop(columns=['Book_Count'])
users_df
```

```python
# setting user_id as index (Not sure if this is necessary or not)
users_df = users_df.rename(columns={'Unnamed: 0': ''}).set_index('')
users_df
```

```python
# Define features set
X = users_df.copy()
X = X.drop("Above_Average", axis=1)
X.head()
```

```python
# Define target vector
y = users_df["Above_Average"].ravel()
y[:5]
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Fit the Random Forest Model

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=64, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

## Make Predictions Using the Testing Data


```python
predictions = rf_model.predict(X_test_scaled)
predictions
```

# Evaluate the Model


```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
