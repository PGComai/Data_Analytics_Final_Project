```python
import pandas as pd

df = pd.read_csv('../CSV/Twice_Cleaned_Country.csv')
df.head()
```

```python
titles = df.groupby('ISBN')
titles.count()
```

```python
top_books = titles.Book_Rating.median()
```

```python
top_books.to_json('../Web/static/js/median_ratings.json')
```

```python
book_ages = titles.Age.median()
book_ages
```

```python
book_ages.to_json('../Web/static/js/median_ages.json')
```

```python
book_info_df = top_books.to_frame(name='median_rating').reset_index()
book_info_df.head()
```

```python
book_ages_df = book_ages.to_frame(name='median_reader_age').reset_index()
book_ages_df.head()
```

```python
book_info_df = book_info_df.merge(book_ages_df)
```

```python
book_instances = titles.User_ID.sum()
book_instances
```

```python
book_inst_df = book_instances.to_frame(name='users_read').reset_index()
book_info_df = book_info_df.merge(book_inst_df)
book_info_df.head()
```
