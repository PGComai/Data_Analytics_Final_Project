```python
import pandas as pd

df = pd.read_csv('../CSV/Cleaned_Country.csv')
df.head()
```

```python
countries = df[['Country', 'User_ID']].groupby('Country')
countries.count()
```

```python
countries.count().to_dict()
```

```python
strip_df = df.copy()
strip_df.Country = strip_df.Country.str.strip()
```

```python
# drop polk, quit, Pender

strip_df.Country = strip_df.Country.str.replace('USA, Guam', 'Guam (USA)')
strip_df.Country = strip_df.Country.str.replace('USA & Canada', 'USA')
strip_df.Country = strip_df.Country.str.replace('America', 'USA')
strip_df.Country = strip_df.Country.str.replace('Australia, Australia', 'Australia')
strip_df.Country = strip_df.Country.str.replace('Catalonia', 'Catalonia (Spain)')
strip_df.Country = strip_df.Country.str.replace('Catalunya, Spain', 'Catalonia (Spain)')
strip_df.Country = strip_df.Country.str.replace('Catalunya Spain', 'Catalonia (Spain)')
strip_df.Country = strip_df.Country.str.replace('Catalunya', 'Catalonia (Spain)')
strip_df.Country = strip_df.Country.str.replace('Deutschland', 'Germany')
strip_df.Country = strip_df.Country.str.replace('Francisco Morazán, Honduras', 'Honduras')
strip_df.Country = strip_df.Country.str.replace('England, united kingdom', 'united kingdom')
strip_df.Country = strip_df.Country.str.replace('Wales, united kingdom', 'united kingdom')
strip_df.Country = strip_df.Country.str.replace('Wales', 'united kingdom')
strip_df.Country = strip_df.Country.str.replace('Tobago', 'Trinidad and Tobago')
# the line below takes care of a problem created by the line above
strip_df.Country = strip_df.Country.str.replace('Trinidad and Trinidad and Tobago', 'Trinidad and Tobago')
strip_df.Country = strip_df.Country.str.replace('galiza', 'Galicia (Spain)')
strip_df.Country = strip_df.Country.str.replace('canary islands, Spain', 'Canary Islands (Spain)')
strip_df.Country = strip_df.Country.str.replace('Venezia Giulia, Italy', 'Venezia Giulia (Italy)')
strip_df.Country = strip_df.Country.str.replace('Orense', 'Galicia (Spain)')
strip_df.Country = strip_df.Country.str.replace('Lleida', 'Catalonia (Spain)')
strip_df.Country = strip_df.Country.str.replace('Italia', 'Italy')
strip_df.Country = strip_df.Country.str.replace('Francisco Morazán, Honduras', 'Honduras')
strip_df.Country = strip_df.Country.str.replace('Euskal herria', 'Basque Country (Spain)')
strip_df.Country = strip_df.Country.str.replace('Bermuda', 'Bermuda (United Kingdom)')
strip_df.Country = strip_df.Country.str.replace('Alachua', 'USA')
strip_df.Country = strip_df.Country.str.replace('Puerto Rico', 'Puerto Rico (USA)')
strip_df.Country = strip_df.Country.str.replace('jersey', 'USA')
strip_df.Country = strip_df.Country.str.replace('united Arab emirates', 'United Arab Emirates', case=True)
strip_df.Country = strip_df.Country.str.replace('United Kingdom', 'united kingdom', case=True)
strip_df.Country = strip_df.Country.str.replace('new Zealand', 'New Zealand', case=True)

strip_df.drop(strip_df.loc[strip_df['Country']=='quit'].index, inplace=True)
strip_df.drop(strip_df.loc[strip_df['Country']=='polk'].index, inplace=True)
strip_df.drop(strip_df.loc[strip_df['Country']=='Pender'].index, inplace=True)

strip_df.Country = strip_df.Country.str.replace('USA', 'United States of America')
strip_df.Country = strip_df.Country.str.title()

strip_df.Country = strip_df.Country.str.replace(' Of ', ' of ', case=True)
strip_df.Country = strip_df.Country.str.replace(' And ', ' and ', case=True)
```

```python
strip_countries = strip_df.groupby('Country')
strip_countries.count()
```
