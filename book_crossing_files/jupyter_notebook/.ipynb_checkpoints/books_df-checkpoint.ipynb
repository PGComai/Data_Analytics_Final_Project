```python
import pandas as pd
import os
```

```python
file = '/Users/nataliebernal/Desktop/git/Data_Analytics_Final_Project/book_crossing_files/original_csv/BX-Books.csv'        

# Determine what the character encoding is
import chardet
with open(file, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
result

# Insert sep to seperate data, character encoding, and skip bad lines
books = pd.read_csv(file, sep=";",encoding='ISO-8859-1', on_bad_lines='skip') 


```

```python
# Display DataFrame
books.head()
```

```python
books.describe()
```

```python
# Drop url column
books = books.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1)
books
```

```python
# Check for duplicates
print(f"Duplicate entries: {books.duplicated().sum()}")
```

```python
for column in books.columns:
    print(f"Column {column} has {books[column].isnull().sum()} null values")
```

```python
books_clean = books.dropna()
```

```python
for column in books_clean.columns:
    print(f"Column {column} has {books_clean[column].isnull().sum()} null values")
    
```

```python
# Print the total number of tradable cryptocurrencies.
print(f"There are {books_clean['ISBN'].unique().size} books")
```

```python
books_clean = books_clean.rename(columns={"ISBN": "isbn", "Book-Title": "book_title", "Book-Author": "book_author", "Year-Of-Publication": "year_of_publication", "Publisher": "publisher"})

books_clean
```

```python
books_clean.set_index('isbn')
```

```python
books_clean.dtypes
```

```python
books_clean = books_clean.replace('"', '', regex=True)
```

```python
# Set the output file name
output_data_file = "/Users/nataliebernal/Desktop/git/Data_Analytics_Final_Project/book_crossing_files/csv/books.csv"
# Export the city_data DataFrame into a CSV file
books_clean.to_csv(output_data_file, index=False)
```
