```python
import pandas as pd
import os
```

```python
file = '/Users/nataliebernal/Desktop/git/Data_Analytics_Final_Project/book_crossing_files/csv/book_crossing_dataset.csv'        

# # Determine what the character encoding is
# import chardet
# with open(file, 'rb') as rawdata:
#     result = chardet.detect(rawdata.read(100000))
# result

# Insert sep to seperate data, character encoding, and skip bad lines
book_crossing_df = pd.read_csv(file, sep=",",encoding='ISO-8859-1', on_bad_lines='skip') 
```

```python
# Check for duplicates
print(f"Duplicate entries: {book_crossing_df.duplicated().sum()}")
```

```python
for column in book_crossing_df.columns:
    print(f"Column {column} has {book_crossing_df[column].isnull().sum()} null values")
```

```python
# Drop Null Values
book_crossing_df = book_crossing_df.dropna()
```

```python
for column in book_crossing_df.columns:
    print(f"Column {column} has {book_crossing_df[column].isnull().sum()} null values")
```

```python
# Remove rows with ages over 100
# we want users that resemble our life expectancy range
book_age = book_crossing_df[(book_crossing_df['age'] > 85) | (book_crossing_df['age'] < 10)].index
```

```python
book_crossing_df.drop(book_age, inplace=True)
book_crossing_df
```

```python
book_crossing_df.sort_values(by=['age'], ascending=False)
```

```python
column_names = ["user_id", "age", "address", "book_rating", "isbn", "book_title", "book_author", "year_of_publication", "publisher"]
book_crossing_df = book_crossing_df.reindex(columns=column_names)
book_crossing_df
```
